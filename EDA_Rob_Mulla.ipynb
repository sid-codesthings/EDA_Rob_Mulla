```python
# Ask a Question about the data

# Lets ask
# What are the locations with the fastest roller coasters(Minimum number of coasters = 10)?
```

```python
df['Location'].value_counts()
```

```python
# We wont take in "Other" location. Let's filter that out.

df.query('Location != "Other"')
```

```python
df.query('Location != "Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean','count'])
```

```python
# To verify...lets verify for "Yomiuriland"
```

```python
df.query('Location == "Yomiuriland"')
```

```python
df.query('Location == "Yomiuriland"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean'])
```

```python
# yes it is right , verified.
```

```python
df.query('Location != "Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean','count'])\
        .query('count >= 10')
```

```python
# lets sort these values by the mean speed

df.query('Location != "Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean','count'])\
        .query('count >= 10')\
        .sort_values('mean') # means sorting by the mean of speed obtained above
```

```python
ax = df.query('Location != "Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean','count'])\
        .query('count >= 10')\
        .sort_values('mean')['mean']\
        .plot(kind='barh',figsize=(12,5),
        title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
